```python
                                     DATA VISULAIZATION  WITH SEABORN 

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

sns.get_dataset_names()

tips = sns.load_dataset('tips')

tips.head()

tips.day.unique() #unique devisions in the dataset

tips.time.unique()

tips.sex.unique()

# UNIVARIATE
"one variable" DATA VISUALIZATION

#STRIP PLOT
sns.stripplot(y='tip',data=tips,jitter=True)# jitter = spread(wide spread) ,no x label 
plt.ylabel('tips($)')
plt.show()

#GROUPING WITH STRIP PLOT
sns.stripplot(x='day',y='tip',data=tips,jitter=True)# the dot indicates the no of peopple who has given the tips
plt.ylabel('tips($)')
plt.show()

sns.stripplot(x='day',y='tip',data=tips,size=5,jitter=False)#size = size of the dots
plt.xlabel('tips($)')
plt.show()

#SWARM PLOT
sns.swarmplot(x='day',y='tip',size=4,data=tips)
plt.xlabel('tips($)')
plt.show()

sns.swarmplot(x='day',y='tip',size=4,hue='sex',data=tips)#hue = sort by color # sat and sunday more tips given by male and female
plt.xlabel('tips($)')
plt.show()

sns.swarmplot(x='day',y='tip',data=tips,hue='sex',size=4)
plt.xlabel('tips($)')
plt.show()

#BOXLOT subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()

import numpy as np
sns.boxplot(x="total_bill", y="day", data=tips, whis=np.inf)
sns.swarmplot(x="total_bill", y="day", data=tips, color=".2")
plt.show()

#violin plot subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()

sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips,
                   color="white", edgecolor="gray")
plt.show()


plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()

#combining plot
sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgrey')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tips($)')
plt.show()

#CATPLOT
#Using catplot() to combine a swarmplot() and a FacetGrid. This allows grouping within additional categorical variables. Using catplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets:
sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="swarm",
                height=4, aspect=.7);

# BIVARIATE 
"two variables" DATA VISUALIZATION

#JOINT plot
sns.jointplot(x='total_bill',y='tip',data=tips)#most of the tips are 2 to 4 if the total bill is between 10 to 24)
plt.show()

#Density plot
sns.jointplot(x='total_bill',y='tip',data=tips,fill=True,kind='kde')
plt.show()

#pair plot
sns.pairplot(tips)
plt.show()

sns.pairplot(tips,hue ='sex')#color coding
plt.show()
```
